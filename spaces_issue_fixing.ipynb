# fixing splited words issue:

## preliminary function

```python
def is_word(word):
    if not wordnet.synsets(word):
        print('is not')
    else:
        print('is')
        
is_word('POL')
is_word('ISHED')
is_word('POLISHED')
```

```python
def remake_words(l=[]):
    i=0
    while i<len(l)-1 and i<15:
        if wordnet.synsets(l[i]+l[i+1]) and ((not wordnet.synsets(l[i])) or (not wordnet.synsets(l[i+1]))):
            l.insert(i, l[i]+l[i+1])
            l.remove(l[i+1])
            l.remove(l[i+1])
            i-=2
        i+=1
    return l
```

```python
def words(s):
    return remake_words(list([t for t in s.replace('"', '" ').split()]))
```

```python
def new_descr(descr):
    return ' '.join(words(descr))
```

### small test

```python
a=df["description"][0]
print(a)
a_words = [t for t in a.replace('"', '" ').split()]
print(a_words)

print(words(a))
print(new_descr(a))
```
